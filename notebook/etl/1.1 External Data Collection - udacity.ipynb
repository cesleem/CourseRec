# 1.1 - External Data Collection - course content from udacity.com

```python
from bs4 import BeautifulSoup
from collections import Counter
import pysrt
import zipfile, glob
import string
import pandas as pd
```

## 0. Unzip Video Subtitles

```python
origin_dir = '../../data/raw/videosubtitles_udacity/archived/'
target_dir = '../../data/raw/videosubtitles_udacity/unarchived/'

print('#### Unzipping course subtitles from /archived to /unarchived ####')

for course_folder in glob.iglob(origin_dir + '*.zip', recursive=False):
    with zipfile.ZipFile(course_folder, "r") as zip_ref:
        course_name = course_folder.replace('../../data/raw/videosubtitles_udacity/archived/', '').replace('Subtitles.zip', '')
        print(course_name)
        zip_ref.extractall(target_dir + '/' + course_name)
```

## 1. By Course Extract Subtitles

```python
print('#### Extracting subtitles from courses ####')

whitelist_alpha = set(' \n\tabcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ')
course_subs = []

for course_folder in glob.iglob(target_dir + '/*', recursive=False):
    print(course_folder)
    course_subs_text = ""
    for sub_path in glob.iglob(course_folder + '/**/*.srt', recursive=True):
        sub_text = pysrt.open(sub_path).text
        course_subs_text += sub_text
        print('\t', sub_text[:10])
        
    #remove non-latin alpha as hack in attempt to remove other languages reduce dataset size
    #TODO - better non-English language removal
    alpha_filtered_course_subs_text = ''.join(filter(lambda x: x in whitelist_alpha, course_subs_text))
    course_subs.append([course_folder.replace('../../data/raw/videosubtitles_udacity/unarchived/', ''), alpha_filtered_course_subs_text])

course_subs_DF = pd.DataFrame(course_subs, columns=['course_slug', 'subtitles'])

course_subs_DF[course_subs_DF.columns] = course_subs_DF.apply(lambda x: x.str.strip())
```
