### 1.1 Get Links

Strategy all li in ul with class "courses", get href

```python
udacityCoursePagePaths = []

#0. Get outline
url = "https://www.udacity.com/georgia-tech"
r = requests.get(url)
soup = BeautifulSoup(r.content)

#1 - start unpacking at 'available courses' list

for li in soup.findAll("ul", {'class':['section-available-courses','courses']})[0].find_all('li'):
    a = li.find('a')
#     print('https://www.udacity.com/{}'.format(a['href']))
    udacityCoursePagePaths.append('https://www.udacity.com/{}'.format(a['href']))
    
#remove courses with 'refresher' in title
udacityCoursePagePaths = [pp for pp in udacityCoursePagePaths if 'refresher' not in pp]
for pp in udacityCoursePagePaths: 
    print(pp) 
```

### 1.2 Get Syllabi for each link

```python
syllabi = {}

for course in course_udacitypage_to_label:
    course_name = course[0]
    print(course_name)
    
    #0. Get udacity course page
    course_url = course[1]
    r = requests.get(course_url)
    soup = BeautifulSoup(r.content)
    syllabus = {}
    if not soup.find("ir-syllabus-list"):
        print("/!\ SORRY COULDN'T FIND ^ \n")
        continue
    for card in soup.find("ir-syllabus-list").findAll("ir-syllabus-card"):
        header = card.select("h2.h4")[0].get_text()
        content = [li.get_text() for li in card.findAll("li")]
        syllabus[header] = content
    syllabi[course_name] = syllabus
```
